```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
bitcoin = pd.read_excel('GDAX.BTC-USD_790301_230316.xlsx')
smp500 = pd.read_csv('^spx_d (1).csv')

bitcoin.drop(range(362),inplace=True)
bitcoin.drop(range(2915, 2990),inplace=True)

bitcoin = bitcoin.iloc[:,[2,7]]
bitcoin.columns = ['Date', 'Close']
bitcoin['Date'] = bitcoin['Date'].map(lambda x:str(x)[0:4]+'-'+str(x)[4:6]+'-'+str(x)[6:8]) 
smp500 = smp500[['Date','Close']]

table = smp500.set_index('Date').join(bitcoin.set_index('Date'), lsuffix='-')
table = table.dropna()
table.columns = ['smp500', 'bitcoin']
table['difference'] = table['bitcoin']-table['smp500']
```

```python
sns.set()
table.dropna(inplace=True)
table.head()
```

```python
#Гистрограмма и график плотности распределения
sns.displot(table['bitcoin'], kde= True)
```

```python
sns.displot(table['bitcoin'], kde= True, bins = 10,  
            line_kws={'lw': 6}, facecolor='lime', edgecolor='black')
```

```python
sns.displot(table[['bitcoin', 'smp500']])
```

```python
#kde_kws - возращение плавующей линии и красим ее красным
# sns.displot(table[['bitcoin', 'smp500']],  
#            kde= False, bins = 10, 
#             line_kws={'lw': 3}, facecolor='lime', edgecolor='black')
```

```python
#kde_kws - возращение плавующей линии и красим ее красным
sns.distplot(table["smp500"],  
            color= "Limegreen", bins = 10,
            hist_kws = {"alpha": 0.75,
                        "edgecolor" : "black"},
             kde_kws = {"color":"red"})
# добавление название графика
plt.xlabel('smp500')
plt.ylabel('Time')
plt.title('Number of Bitcoin trades per hour')
```

```python
sns.distplot(table["smp500"], color= "Limegreen",kde= False,
            bins = 100,
            hist_kws = {"alpha": 0.75,
                        "edgecolor" : "black"},
             kde_kws = {"color":"red"})
plt.xlabel('smp500')
plt.ylabel('Time')
plt.title('Number of Bitcoin trades per hour')
# метод vlines
# вертикальная линия отмечает средняя значение по столбцу
plt.vlines(x = table["smp500"].mean(), ymin=0, ymax= 60,colors = "red", linestyles= "dotted")
```

```python
# функиция kdeplot - сглаживания графика
sns.kdeplot(table[["smp500",'bitcoin']], shade= True)
plt.xlabel('smp500')
plt.ylabel('Time')
plt.title('Number of Bitcoin trades per hour')
```

```python
# «выключаем» гистограмму
# «включаем» rugs

sns.displot(table["smp500"],rug=True, kind='kde', color = 'red')
plt.xlabel("Age")
plt.ylabel("Counts")
plt.title("Age of responsents")
```
